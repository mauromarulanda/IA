```python
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('./Salary Data.csv')
data
```

```python
linear_model = smf.ols(formula = 'Age ~ Salary', data = data).fit()
```

```python
linear_model.params
```

```python
linear_model.summary()
```

```python
salary_pred = linear_model.predict(pd.DataFrame(data['Salary']))
salary_pred
```

```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
import numpy as np

# Cargar los datos desde un archivo CSV y eliminar filas con valores faltantes
datos = pd.read_csv("./Salary Data.csv").dropna()

age = datos["Age"].values.reshape((-1, 1))
salary = datos["Salary"]

regresion = linear_model.LinearRegression()

modelo = regresion.fit(age, salary)

print("IntersecciÃ³n (b)", modelo.intercept_)
print("Pendiente (m)", modelo.coef_)

entrada = [[25],[30], [35], [40], [45], [50]]
prediccion = modelo.predict(entrada)
print("Predicciones:", prediccion)

plt.ylabel("Salary")
plt.xlabel("Age")
plt.scatter(age, salary, color="blue", alpha=0.55)
plt.scatter(entrada, modelo.predict(entrada), color="red")

# Trazar la lÃ­nea de regresiÃ³n lineal
plt.plot(age, modelo.predict(age), color="red", linewidth=3)

plt.show()

```
